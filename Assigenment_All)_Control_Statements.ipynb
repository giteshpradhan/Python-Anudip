1. Python program to check leap year

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

4. A toy vendor supplies three types of toys: Battery Based Toys, Key-based
Toys, and Electrical Charging Based Toys. The vendor gives a discount of
10% on orders for battery-based toys if the order is for more than Rs. 1000.
On orders of more than Rs. 100 for key-based toys, a discount of 5% is
given, and a discount of 10% is given on orders for electrical charging based
toys of value more than Rs. 500. Assume that the numeric codes 1,2 and 3
are used for battery based toys, key-based toys, and electrical charging based
toys respectively. Write a program that reads the product code and the order
amount and prints out the net amount that the customer is required to pay
after the discount.

```python
code = int(input("Enter product code (1-Battery, 2-Key, 3-Electrical): "))
amount = float(input("Enter order amount (Rs.): "))
discount = 0
if code == 1:  # Battery Based Toys
    if amount > 1000:
        discount = 0.10 * amount
elif code == 2:  # Key-based Toys
    if amount > 100:
        discount = 0.05 * amount
elif code == 3:  # Electrical Charging Based Toys
    if amount > 500:
        discount = 0.10 * amount
else:
    print("Invalid product code!")
net_amount = amount - discount
print("Net amount to be paid: Rs. {net_amount}")
```

1. Write a python program to reverse a number using a while loop.

```python
num = int(input("Enter a number: "))
reverse_num = 0

while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num = num // 10

print("Reversed number:", reverse_num)
```

2. Write a python program to check whether a number is palindrome or
not?

```python

def is_palindrome(num):
    str_num = str(num)
    return str_num == str_num[::-1]
number = int(input("Enter a number: "))
if is_palindrome(number):
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")
```

3. Write a python program finding the factorial of a given number using
a while loop.

```python

num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"Factorial of {num} is {factorial}")
```

4. Accept numbers using input() function until the user enters 0. If user
input 0 then break the while loop and display the sum of all the
numbers.

```python
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break
    total += num

print("The sum of all numbers is: {total}")
```

2. Python program to check if the given string is a palindrome

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]
string = input("Enter a string: ")

if is_palindrome(string):
    print('"{string}" is a palindrome.')
else:
    print('"{string}" is not a palindrome.')
```

3. Python program to check if a given number is an Armstrong number

```python

num = int(input("Enter a number: "))
order = len(str(num))
sum_of_powers = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** order
    temp //= 10

if num == sum_of_powers:
    print("{num} is an Armstrong number.")
else:
    print("{num} is not an Armstrong number.")
```

4. Python program to get the Fibonacci series between 0 to 50


```python
a, b = 0, 1

print("Fibonacci series between 0 and 50:")

while a <= 50:
    print(a, end=" ")
    a, b = b, a + b
```

5. Python program to check the validity of password input by users


```python
import re

password = input("Enter your password: ")
if (6 <= len(password) <= 16
    and re.search("[a-z]", password)
    and re.search("[A-Z]", password)
    and re.search("[0-9]", password)
    and re.search("[$#@]", password)):
    print("Valid Password")
else:
    print("Invalid Password")
```
