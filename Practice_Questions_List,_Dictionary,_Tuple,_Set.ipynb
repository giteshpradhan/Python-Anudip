14. Write a program to check whether a key exists in a dictionary.

```python
student = {
    "Name": "Rahul Sharma",
    "Age": 20,
    "Roll Number": "CS101",
    "Course": "Computer Science"
}

key_to_check = "Age"

if key_to_check in student:
    print("Key '{key_to_check}' exists in the dictionary with value: {student[key_to_check]}")
else:
    print("Key '{key_to_check}' does not exist in the dictionary.")
```

18. Write a program to remove an element from a set.

```python

my_set = {10, 20, 30, 40, 50}

print("Original Set:", my_set)
my_set.remove(30)

print("Set after removing 30:", my_set)

my_set.discard(50)

print("Set after discarding 50:", my_set)
try:
    my_set.remove(100)
except KeyError as e:
    print("Error occurred while removing 100:", e)

my_set.discard(100)
print("Set after discarding 100:", my_set)
```

19. Write a program to find the union of two sets.

```python

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

print("Set 1:", set1)
print("Set 2:", set2)

union_set = set1.union(set2)

print("\nUnion of Set 1 and Set 2:", union_set)

union_set_alt = set1 | set2
print("Union using | operator:", union_set_alt)
```
