Write a Python program to create a NumPy array with five integers and display it.



```python
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50])
print("NumPy Array:", arr)
```

Write a program to find the data type of a NumPy array.

```python
arr = np.array([10, 20, 30, 40, 50])
print("NumPy Array:", arr)
print("Data type of array:", arr.dtype)
```

Write a program to access the first and last element of a NumPy array.

```python
arr = np.array([10, 20, 30, 40, 50])
first_element = arr[0]
last_element = arr[-1]
print("NumPy Array:", arr)
print("First element:", first_element)
print("Last element:", last_element)
```

Write a program to slice a NumPy array and print elements from index 1 to 3.

```python

arr = np.array([10, 20, 30, 40, 50])
sliced_arr = arr[1:4]
print("Original array:", arr)
print("Sliced elements (index 1 to 3):", sliced_arr)
```

Write a program to display the first three elements of a NumPy array using slicing.


```python
arr = np.array([5, 10, 15, 20, 25])
first_three = arr[:3]
print("Original array:", arr)
print("First three elements:", first_three)
```
