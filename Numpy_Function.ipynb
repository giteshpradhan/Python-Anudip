1.You are given a NumPy array of student marks. Write a program to extract marks greater than the class average.




```python
import numpy as np
marks = np.array([35, 80, 55, 20, 90, 45, 70])
class_avg = np.mean(marks)
above_avg = marks[marks > class_avg]

print("Class Average:", class_avg)
print("Marks above average:", above_avg)
```

2.A company stores daily sales data in a NumPy array. Write a program to find days where sales were higher than 50,000.




```python
import numpy as np
sales = np.array([40000, 52000, 61000, 30000, 75000, 48000, 90000])
high_sales_days = np.where(sales > 50000)[0]
high_sales_values = sales[sales > 50000]
print("Days with sales > 50,000:", high_sales_days)
print("Sales values on those days:", high_sales_values)
```

3.Given a NumPy array of temperatures in Celsius, write a program to convert all values to Fahrenheit using vectorized operations.


```python

celsius = np.array([0, 20, 37, 100])
fahrenheit = celsius * 9/5 + 32

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

5.Given a NumPy array of product prices, write a program to apply a 10 percent discount on prices greater than 1,000.


```python

prices = np.array([500, 1200, 800, 2500, 1500, 950])
discounted_prices = np.where(prices > 1000, prices * 0.9, prices)

print("Original Prices:", prices)
print("Discounted Prices:", discounted_prices)
```

9*.A 2D NumPy array represents sales data of products across different regions. Write a program to find the region with the highest total sales.




```python
import numpy as np
sales = np.array([
    [20000, 30000, 25000],
    [40000, 35000, 30000],
    [15000, 20000, 18000],
    [50000, 45000, 40000]
])

region_totals = np.sum(sales, axis=1)
highest_region_index = np.argmax(region_totals)
highest_region_sales = region_totals[highest_region_index]

print("Region totals:", region_totals)
print("Region with highest sales:", highest_region_index + 1)
print("Highest total sales:", highest_region_sales)
```

10.Given a NumPy array of sensor readings, write a program to detect and extract all abnormal readings that are more than 2 times the mean value.


```python
readings = np.array([10, 15, 20, 100, 25, 200, 30])
mean_value = np.mean(readings)
abnormal = readings[readings > 2 * mean_value]

print("Mean value:", mean_value)
print("Abnormal readings:", abnormal)
```

11.A NumPy array contains salary data of employees. Write a program to apply tax rules:


Salary above 50,000 → deduct 10 percent


Salary above 1,00,000 → deduct 20 percent


```python
salaries = np.array([40000, 60000, 120000, 80000, 30000, 150000])
net_salaries = np.where(
    salaries > 100000, salaries * 0.8,
    np.where(salaries > 50000, salaries * 0.9, salaries))

print("Original Salaries:", salaries)
print("Net Salaries after Tax:", net_salaries)

```

14.Given a NumPy array of daily profit and loss values, write a program to separate profit days and loss days using boolean indexing.




```python
import numpy as np
profits_losses = np.array([200, -150, 300, -50, 400, -100, 250])
profit_days = profits_losses[profits_losses > 0]
loss_days = profits_losses[profits_losses < 0]

print("Profit days:", profit_days)
print("Loss days:", loss_days)
```

16.A NumPy array contains ratings given by users. Write a program to normalize the ratings between 0 and 1.




```python
import numpy as np
ratings = np.array([2, 5, 7, 10, 3, 8])

normalized_ratings = (ratings - np.min(ratings)) / (np.max(ratings) - np.min(ratings))

print("Original Ratings:", ratings)
print("Normalized Ratings:", normalized_ratings)
```

17.Given a NumPy array of transaction amounts, write a program to flag suspicious transactions that are either extremely low or extremely high.


```python
import numpy as np
transactions = np.array([50, 2000, 50000, 5, 120000, 800, 300000])
low_threshold = 100
high_threshold = 100000
suspicious = transactions[(transactions < low_threshold) | (transactions > high_threshold)]

print("Suspicious transactions:", suspicious)
```

18.A 2D NumPy array stores marks of students across tests. Write a program to identify the topper based on total marks.




```python

marks = np.array([
    [80, 75, 90],
    [60, 70, 65],
    [95, 85, 100],
    [40, 50, 55]
])
student_totals = np.sum(marks, axis=1)
topper_index = np.argmax(student_totals)
topper_total = student_totals[topper_index]

print("Total marks of each student:", student_totals)
print("Topper is Student", topper_index + 1, "with total marks:", topper_total)
```
