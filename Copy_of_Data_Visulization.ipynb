Type of Graph

```python
import matplotlib.pyplot as plt
```

```python
month =[1,2,3,4,5]
sales=[2000,3000,4000,6000,9000]
#Line chart
plt.plot(month,sales)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly sales")
plt.show()
```

```python
product =["A","B","C"]
sales =[300,4500,6000]
#Bar Chart
plt.bar(product,sales)
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product sales")
plt.show()
```

```python
marks =[35,45,50,60,75,80,85,95,99]
#Histogram
plt.hist(marks,bins=5)# In histogram graph we can show the range (Bins show the range of graph)
plt.xlabel("marks Range")
plt.ylabel("No of Students")
plt.title("Marks distribution")
plt.show()
```
