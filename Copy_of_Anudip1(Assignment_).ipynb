9. Write a program to check whether a year is a leap year or not.


```python


year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

10. Write a program to calculate bonus for an employee based on salary.

```python
salary = float(input("Enter employee salary: "))

if salary <= 10000:
    bonus = salary * 0.20
elif salary <= 50000:
    bonus = salary * 0.15
else:
    bonus = salary * 0.10

print("Salary:", salary)
print("Bonus:", bonus)
print("Total Pay (Salary + Bonus):", salary + bonus)
```

15. Write a program to count the number of digits in a given number.

```python
num = int(input("Enter a number: "))
count = 0


if num == 0:
    count = 1
else:
    while num != 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```

19. Write a program to take marks of 5 students and calculate the average

```python
total = 0
for i in range(1, 6):
    marks = float(input(f"Enter marks of student {i}: "))
    total += marks

average = total / 5
print("Average marks of 5 students:", average)

```

20. Write a program to take 5 numbers from the user and count positive and negative numbers.

```python
positive_count = 0
negative_count = 0

for i in range(1, 6):
    num = int(input(f"Enter number {i}: "))
    if num > 0:
        positive_count += 1
    elif num < 0:
        negative_count += 1

print("Total Positive numbers:", positive_count)
print("Total Negative numbers:", negative_count)
```
